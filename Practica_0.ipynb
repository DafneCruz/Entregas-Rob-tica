## Previo de la Práctica 0

### La estructura de paquetes en ROS 2.
En ROS 2, los paquetes son la base para organizar el código. Dentro de un paquete se agrupan todos los elementos necesarios para desarrollar una aplicación robótica, como nodos, configuraciones y dependencias.

Un paquete en Python generalmente incluye:

- package.xml: Contiene la información general del paquete, como su nombre, versión, dependencias y licencia.

- setup.py: Se utiliza para configurar e instalar el paquete, además de registrar los nodos que se podrán ejecutar.

- resource/: Carpeta que guarda un archivo con el nombre del paquete, lo cual permite que ROS 2 lo identifique correctamente.

- Carpeta principal del paquete (por ejemplo, p0_py/): Aquí se encuentra el código fuente, incluyendo los nodos programados en Python.

Esta organización facilita el desarrollo, mantenimiento y reutilización del código dentro de proyectos robóticos.

### Definición de Nodos de Tópicos.
En ROS 2, un nodo es un programa que forma parte de un sistema robótico y se encarga de realizar una tarea específica, como procesar datos, controlar dispositivos o generar información. Cada nodo funciona de manera independiente, pero puede comunicarse con otros para formar sistemas más complejos. Esta comunicación se realiza mediante tópicos, los cuales actúan como canales a través de los cuales los nodos intercambian mensajes. En este esquema, un nodo puede desempeñar el papel de publicador, enviando datos a un tópico, mientras que otro nodo puede ser suscriptor, recibiendo y procesando dicha información. Este tipo de comunicación no requiere que los nodos estén directamente conectados entre sí.

La comunicación en ROS 2 se basa en el modelo publicador–suscriptor, donde un nodo publica información de manera continua y uno o varios nodos pueden suscribirse para recibir esos datos. Este modelo permite que los sistemas sean más flexibles, modulares y fáciles de escalar. En el ámbito de la robótica, el uso de nodos y tópicos es fundamental, ya que facilita dividir el sistema en partes más pequeñas y manejables, reutilizar código en distintas aplicaciones, mejorar la comunicación entre componentes y desarrollar sistemas distribuidos de manera eficiente.

## Resultados

Durante la realización de la práctica se logró implementar correctamente la comunicación entre nodos utilizando ROS 2. Se programó un nodo publicador encargado de generar una señal senoidal utilizando un mensaje de tipo `Float64`, cuyos valores varían en un rango de 0 a 1. Esta señal representa la velocidad en revoluciones por minuto (RPM) y es enviada a través del tópico `/rpm_topic`.

Posteriormente, se desarrolló un nodo suscriptor-publicador que recibe los valores de RPM y realiza la conversión a radianes por segundo (rad/s). Esta conversión se lleva a cabo mediante la relación:

$$
\omega = \frac{2\pi}{60} \cdot RPM
$$

El resultado es publicado en el tópico `/rad_s_topic`, permitiendo visualizar la transformación de la señal en tiempo real.

Se verificó el correcto funcionamiento del sistema mediante el uso de comandos como `ros2 topic echo`, observando que los valores generados siguen un comportamiento periódico y continuo.

## Gráficas

Para analizar el comportamiento de la señal generada, se realizó una gráfica de la señal senoidal en función del tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tiempo
t = np.linspace(0, 10, 100)

# Señal senoidal (RPM)
rpm = 0.5 * (np.sin(t) + 1)

# Conversión a rad/s
rad_s = (2 * np.pi / 60) * rpm

# Graficar
plt.plot(t, rpm, label="RPM")
plt.plot(t, rad_s, label="rad/s")

plt.xlabel("Tiempo (s)")
plt.ylabel("Valor")
plt.title("Señal senoidal en RPM y su conversión a rad/s")
plt.legend()

plt.show()
```

### Resultados Númericos

```python
import numpy as np

rpm_max = 1
rpm_min = 0

omega_max = (2 * np.pi / 60) * rpm_max
omega_min = (2 * np.pi / 60) * rpm_min

print("RPM max:", rpm_max)
print("RPM min:", rpm_min)
print("ω max (rad/s):", omega_max)
print("ω min (rad/s):", omega_min)
```

Estos resultados confirman que la señal convertida mantiene el mismo comportamiento senoidal, pero expresado en unidades de radianes por segundo.
